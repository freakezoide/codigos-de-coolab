```python
import pandas as pd
import requests
import json
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# URL del JSON (reemplaza con la correcta)
url = "https://observatorio.ceibal.edu.uy/static/assets/spa.541157ad193384bd551f.entry.js"

# Headers para simular un navegador (opcional, si el servidor lo requiere)
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

try:
    # Realizar la petición HTTPS (sin verificar SSL)
    response = requests.get(url, headers=headers, verify=False)
    response.raise_for_status()  # Lanza error si la respuesta no es 200 OK

    # Intentar cargar el JSON
    data = response.json()

    # Convertir a DataFrame y guardar como CSV
    df = pd.DataFrame(data)
    df.to_csv("datos_ceibal.csv", index=False)
    print("¡Archivo CSV guardado correctamente!")

    # Descargar el CSV desde Colab
    from google.colab import files
    files.download("datos_ceibal.csv")

except Exception as e:
    print(f"Error: {e}")
    print("Respuesta del servidor:", response.text)  # Ver contenido crudo de la respuesta
```
